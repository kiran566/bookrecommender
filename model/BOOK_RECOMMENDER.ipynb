```python
# merging  the dataset books and ratings in ISBN
books_ratings=books.merge(ratings,on='ISBN')
```

```python
books_ratings
```

```python
books_ratings.dtypes
```

```python
# rating count for each title
rating_count=books_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
rating_count.rename(columns={'Book-Rating':'rating_count'},inplace=True)
rating_count
```

```python
# average rating
average_rating=books_ratings.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
average_rating.rename(columns={'Book-Rating':'average_rating'},inplace=True)
average_rating
```

popular movies means===="the movie having more ratings rovided wtih an average rating

```python
# mergint the 2 datasets
popular_movies=rating_count.merge(average_rating,on='Book-Title')
popular_movies
```

```python
# meaningfully
popular_movies=popular_movies[popular_movies['rating_count']>=250].sort_values('average_rating',ascending=False).head(50)
```

```python
popular_movies
```

```python
popular_movies.drop_duplicates(inplace=True)
popular_movies.shape
```

```python
# to get the url of those images and all
popular_movies=popular_movies.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','rating_count','average_rating']]
```

=============COLLABERATIVE FILTERING=======================================================

```python
# USRS WHO RATED BOOKS MORE THAN 200 TIMES
x=books_ratings.groupby('User-ID').count()['Book-Rating']>200
x[x]
# only those whoe rated more than 200 times
```

```python
user_indexes=x[x].index
user_indexes
```

```python
# those total entries
con=books_ratings['User-ID'].isin(user_indexes)
con
```

```python
filterd_rating=books_ratings[con]
filterd_rating
```

```python
# intis also take only those books which are rated more than 50 times
y=filterd_rating.groupby('Book-Title').count()['Book-Rating']>=50
y[y]
```

```python
# then
famous_books=y[y].index
famous_books
```

```python
cony=filterd_rating['Book-Title'].isin(famous_books)
cony
```

```python
final_books=filterd_rating[cony]
final_books
```

```python
# making pivot table
# as books title on rows and users on columns
pt=final_books.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
# hence each book is an vector of ratings book=[.......810]
# can use simiiratyscores
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores.shape  #each movie with itself=1 and other movies similiarty
```

```python
np.where(pt.index=="4 Blondes")[0][0]
```

```python
def book_recommend(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6] #top5
  for i in similar_items:
    print(pt.index[i[0]])
```

```python
book_recommend("Message in a Bottle")
```
